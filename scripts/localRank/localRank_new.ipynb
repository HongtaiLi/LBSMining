```python
# 214418658    1851
# 109307439    1257
# 338901468    1062
# 14372827      961
# 38209268      803
# 36543482      692
# 37610007      652
# 455222975     618
# 203730976     584
# 558163359     581
# 260528472     565
# 20596121      507
# 14936915      499
# 24822574      493
# 68548440      486


#user_cat_mat.reset_index()
%pylab inline

import sys
import urllib2
import time
import json
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder


df = pd.read_csv('./data/check_in_nyc.csv')
venue_to_super = df[['v_id','super_cat']]
venue_to_lat = df[['v_id','lat']]
venue_to_lon = df[['v_id','lon']]
venue_count = None

#venue_to_lat[venue_to_lat.v_id=='4c9b818cd4b1b1f7927bd935']['lat'].unique()[0]
super_cat_arr = venue_to_super['super_cat'].unique()


print super_cat_arr
#print np.where(super_cat_arr=='Residence')[0]
#print venue_to_super[venue_to_super.v_id=='4c9b818cd4b1b1f7927bd935']['super_cat'].unique()[0]

def url_request(venue_id,f_out):
    url_str = "https://api.foursquare.com/v2/venues/venue_id?oauth_token=UINIO453V10B5B22RB33ODC3SNKBMUZTT5KLPMUCUAYAHFDR&v=20140317"
    url_str_new = url_str.replace('venue_id',venue_id.strip())
    
    response = urllib2.urlopen(url_str_new)
    remaining = response.info()['X-RateLimit-Remaining']
    content = response.read()
    print type(remaining)
    if int(remaining) < 10:
        time.sleep(60*6)
    
    out_line = venue_id.strip() + '\t' + content.strip()+'\n'
    f_out.write(out_line)
    f_out.flush()

    print remaining

def craw(v_list,out):
        
    f_out = open(out,'a')
    
    for venue_id in v_list:
        url_request(venue_id,f_out)
        
    f_out.close()



def get_sub_venues(user_id):
    df_userindex = df.set_index('user_id')
    checkin_subgroup = df_userindex.loc[user_id]
    venue_count = checkin_subgroup['v_id'].value_counts()
    v_list = list(venue_count.index)
    return v_list,venue_count

def craw_venues(v_list,out):
    craw(v_list,out)



def extract_feature(row):
    try:
        content = row['content']
        #print content
        v_id = row['id']
        v_count = venue_count.ix[v_id]
        decoded_content = json.loads(content.strip())
    except:
        print sys.exc_info()[0]
        return pd.Series({})
        
    
    check_in_count = 0
    unique_user_count = 0
    num_of_likes = 0
    pcode = 0
    rating = 0
    cat_index = 8
    lat = 0.0
    lon = 0.0
    photo_counts = 0
    price_tier = 0
    tip_count = 0
    verified = 0
    
    try:
        super_cat = venue_to_super[venue_to_super.v_id==v_id]['super_cat'].unique()[0]
        cat_index = np.where(super_cat_arr==super_cat)[0][0]
    except:
        cat_index = 8
    
    try:
        #lat = venue_to_lat[venue_to_lat.v_id==v_id]['lat'].unique()[0]
        #lon = venue_to_lon[venue_to_lon.v_id==v_id]['lon'].unique()[0]
        lat = decoded_content['response']['venue']['location']['lat']
        lon = decoded_content['response']['venue']['location']['lng']
        
    except:
        lat = 0.0
        lon = 0.0
    
    try:
        check_in_count =  int(decoded_content['response']['venue']['stats']['checkinsCount'])
    except:
        check_in_count = 0
    
    try:
        unique_user_count = int(decoded_content['response']['venue']['stats']['usersCount'])
    except:
        unique_user_count = 0
    
    try:
        tip_count = int(decoded_content['response']['venue']['stats']['tipCount'])
    except:
        tip_count = 0
        
        
    try:
        num_of_likes =  int(decoded_content['response']['venue']['likes']['count'])
    except:
        num_of_likes = 0
        
        
    try:
        pcode =  int(decoded_content['response']['venue']['location']['postalCode'])
    except:
        pcode = 0
                
    try:
        rating = float(decoded_content['response']['venue']['rating'])
    except:
       raiting = 0
    
    try:
        photo_counts = int(decoded_content['response']['venue']['photos']['count'])
    except:
        photo_counts = 0
    
    try:
        price_tier = int(decoded_content['response']['venue']['price']['tier'])
    except:
        price_tier = 0
    
    try:
        verified_flag = bool(decoded_content['response']['venue']['verified'])
        
        if verified_flag:
            verified = 1
        else:
            verified = 0
            
            
    except:
        verified = 0
        
        
    
    #print check_in_count,unique_user_count,num_of_likes,pcode,cat_index,lat,lon
    
    return pd.Series({'v_id':v_id,'checkinsCount':check_in_count,'usersCount':unique_user_count,'tip_count':tip_count,
                      'likes':num_of_likes,'photo_counts':photo_counts,'rating':rating,'v_count':v_count,'postalCode':pcode,
                      'categories':cat_index,'lat':lat,'lon':lon,'price_tier':price_tier,'verified':verified})
#'postalCode':pcode


def get_venue_feature_matrix(venus_path,user_id):
    global venue_count
    #venue_group_path = './group_venues_out4'

    v_list,venue_count = get_sub_venues(user_id)
    
    df_venue_info = pd.read_csv(venus_path,sep='\t',names=['id','content'])
    #print df_venue_info
    
    df_venue_feature = df_venue_info.apply(extract_feature,axis=1)
    df_venue_feature = df_venue_feature.dropna()
    df_venue_feature = df_venue_feature.set_index('v_id')
    #print shape(df_venue_feature)
    return df_venue_feature


# print df_venue_feature.head()
    
import itertools
import numpy as np
 
from sklearn import svm, linear_model, cross_validation
from sklearn import preprocessing

 
 
def transform_pairwise(X, y):
    """Transforms data into pairs with balanced labels for ranking
 
    Transforms a n-class ranking problem into a two-class classification
    problem. Subclasses implementing particular strategies for choosing
    pairs should override this method.
 
    In this method, all pairs are choosen, except for those that have the
    same target value. The output is an array of balanced classes, i.e.
    there are the same number of -1 as +1
 
    Parameters
    ----------
    X : array, shape (n_samples, n_features)
        The data
    y : array, shape (n_samples,) or (n_samples, 2)
        Target labels. If it's a 2D array, the second column represents
        the grouping of samples, i.e., samples with different groups will
        not be considered.
 
    Returns
    -------
    X_trans : array, shape (k, n_feaures)
        Data as pairs
    y_trans : array, shape (k,)
        Output class labels, where classes have values {-1, +1}
    """
    X_new = []
    y_new = []
    y = np.asarray(y)
    if y.ndim == 1:
        y = np.c_[y, np.ones(y.shape[0])]
    comb = itertools.combinations(range(X.shape[0]), 2)
    for k, (i, j) in enumerate(comb):
        if y[i, 0] == y[j, 0] or y[i, 1] != y[j, 1]:
            # skip if same target or different group
            continue
        X_new.append(X[i] - X[j])
        y_new.append(np.sign(y[i, 0] - y[j, 0]))
        # output balanced classes
        if y_new[-1] != (-1) ** k:
            y_new[-1] = - y_new[-1]
            X_new[-1] = - X_new[-1]
    return np.asarray(X_new), np.asarray(y_new).ravel()
 
 
class RankSVM(svm.LinearSVC):
    """Performs pairwise ranking with an underlying LinearSVC model
 
    Input should be a n-class ranking problem, this object will convert it
    into a two-class classification problem, a setting known as
    `pairwise ranking`.
 
    See object :ref:`svm.LinearSVC` for a full description of parameters.
    """
 
    def fit(self, X, y):
        """
        Fit a pairwise ranking model.
 
        Parameters
        ----------
        X : array, shape (n_samples, n_features)
        y : array, shape (n_samples,) or (n_samples, 2)
 
        Returns
        -------
        self
        """
        X_trans, y_trans = transform_pairwise(X, y)
        super(RankSVM, self).fit(X_trans, y_trans)
        return self
 
    def predict(self, X):
        """
        Predict an ordering on X. For a list of n samples, this method
        returns a list from 0 to n-1 with the relative order of the rows of X.
 
        Parameters
        ----------
        X : array, shape (n_samples, n_features)
 
        Returns
        -------
        ord : array, shape (n_samples,)
            Returns a list of integers representing the relative order of
            the rows in X.
        """
        if hasattr(self, 'coef_'):
            np.argsort(np.dot(X, self.coef_.T))
        else:
            raise ValueError("Must call fit() prior to predict()")
 
    def score(self, X, y):
        """
        Because we transformed into a pairwise problem, chance level is at 0.5
        """
        X_trans, y_trans = transform_pairwise(X, y)
        return np.mean(super(RankSVM, self).predict(X_trans) == y_trans)

def get_df_venue_feature(venues_path,user_id):
    return get_venue_feature_matrix(venues_path,user_id)


def run_local_rank_svm(venues_path,user_id):
    #df_venue_feature = get_venue_feature_matrix(venue_group_path)
    feature_list = ['checkinsCount','likes','rating','usersCount','tip_count','photo_counts','categories','price_tier']
    df_venue_feature = get_df_venue_feature(venues_path,user_id)
   
   # categ = [feature_list.index(x) for x in 'rating', 'categories', 'price_tier']
    #enc = OneHotEncoder(categorical_features = categ)
    
    
    X = df_venue_feature[feature_list].values #,'postalCode','lat','lon'
    #X = enc.fit_transform(df_venue_feature[feature_list]).toarray()
    Y  = df_venue_feature['v_count'].values
    #print X.shape
    #print 'test:',Y
    #Y = preprocessing.scale(Y)
    
    n_samples, n_features = df_venue_feature.shape
    n_features -= 1
    cv = cross_validation.KFold(n_samples, 4)
    train, test = iter(cv).next()
 

    # print the performance of ranking
    
    rank_svm = RankSVM().fit(X[train], Y[train])
    print rank_svm.coef_
    
    print 'Performance of ranking ', rank_svm.score(X[test], Y[test])
    
    ridge = linear_model.RidgeCV(fit_intercept=True)
    ridge.fit(X[train], Y[train])
    X_test_trans, y_test_trans = transform_pairwise(X[test], Y[test])
    score = np.mean(np.sign(np.dot(X_test_trans, ridge.coef_)) == y_test_trans)
    
    print ridge.coef_
    print 'Performance of linear regression ', score

    return feature_list,ridge.coef_ 

# def run_linear_regression(venues_path):
    

def plot_weight(feature_list,coef):
    fig = plt.figure()
    ax = plt.subplot(111)
    width=1.0
    #print len(feature_list),len(coef[0])
    ax.bar(np.arange(len(feature_list)), coef, width=width)
    ax.set_xticks(np.arange(len(feature_list)))
    ax.set_xticklabels(feature_list, rotation=90)
    plt.show()
    
def get_sub_rank(venues_path,user_id):
    feature_list,coef = run_local_rank_svm(venues_path,user_id)
    plot_weight(feature_list,coef)
    
def craw_venue_by_userid(user_id,v_path):
    v_list,v_count = get_sub_venues(user_id)
    craw_venues(v_list,v_path)
    print 'done'
        
if __name__ == '__main__':
    # as showcase, we will create some non-linear data
    # and print the performance of ranking vs linear regression
 
    #craw_venue_by_userid(109307439,'109307439_venues')
    #df_venue_feature = get_venue_feature_matrix(venues_path)


    get_sub_rank('214418658_venues',214418658)
    get_sub_rank('109307439_venues',109307439)
    get_sub_rank('37610007_venues',37610007)
    
    # and that of linear regression
#     ridge = linear_model.RidgeCV(fit_intercept=True)
#     ridge.fit(X[train], Y[train])
#     X_test_trans, y_test_trans = transform_pairwise(X[test], Y[test])
#     score = np.mean(np.sign(np.dot(X_test_trans, ridge.coef_)) == y_test_trans)
#     print 'Performance of linear regression ', score
    
```
